- acessar planilha e extrair as colunas pra conferencia 

```python
import pandas as pd

df_composicao = pd.read_excel('Composição 151.xlsx',header=3)

#print(df_composicao.columns)
df_resumido = df_composicao[['DATA','DESCRIÇÃO DA CONTA.1','DÉBITO (+)','CRÉDITO (+)']]
 


```

- Preciso agrupar por data e descrição da conta.1 e somar os valores da coluna crédito 



```python
#converter para numero 
df_composicao['CRÉDITO (+)'] = pd.to_numeric(df_composicao['CRÉDITO (+)'], errors='coerce')

df_credito = (
    df_composicao[['DATA','DESCRIÇÃO DA CONTA.1','DÉBITO (+)','CRÉDITO (+)']]
    .groupby(['DATA','DESCRIÇÃO DA CONTA.1'], as_index=False)
    .sum()
)

#print(df_composicao.columns)
#df_credito
```

- Validar se os valores de credito que foram somados batem com a coluna debito e marcar com a linha amarela  

```python
import pandas as pd

df_composicao['DÉBITO (+)'] = pd.to_numeric(df_composicao['DÉBITO (+)'], errors='coerce')
df_credito['CRÉDITO (+)'] = pd.to_numeric(df_credito['CRÉDITO (+)'], errors='coerce')

df_composicao['CONFERE'] = df_composicao['DÉBITO (+)'].isin(df_credito['CRÉDITO (+)'])

writer = pd.ExcelWriter('resultado_conferencia.xlsx', engine='xlsxwriter')

df_composicao.to_excel(writer, index=False, sheet_name='resultado')

workbook = writer.book
worksheet = writer.sheets['resultado']

formato_amarelo = workbook.add_format({'bg_color': 'yellow'})

for i, valor in enumerate(df_composicao['CONFERE'], start=1):
    if valor:
        worksheet.set_row(i, None, formato_amarelo)

writer.close()
```
